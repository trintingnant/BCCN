```python
def m_func(t):
    
    return .96 * (1 - np.exp(-t/ .25))

np.vectorize(m_func)

def h_func(t):
    
    return .6 * np.exp(-t/.84)

np.vectorize(h_func)

def n_func(t):
    
    return .89 - .38 * np.exp(-t/1.7)

np.vectorize(n_func)

gNA_new = 70.7
gk_new = 24.3

timecourse = np.arange(0,10,.1)

plt.figure(figsize = (15,10))

plt.plot(timecourse, n_func(timecourse), label = "n")
plt.plot(timecourse, m_func(timecourse), label = "m")
plt.plot(timecourse, h_func(timecourse), label = "h")
plt.title('Timecourse of Ion channel Gating variables')
plt.xlabel('Time in $ms$')
plt.ylabel('Percentage of open channels')
plt.legend()


```

```python

def gNA(t):
    
    return 70.7 * np.power(m_func(t),3) * h_func(t)

np.vectorize(gNA)

def gK(t):
    
    return 24.3 * np.power(n_func(t),4)

np.vectorize(gK)

plt.figure()

plt.figure(figsize = (15,10))

plt.plot(timecourse, gNA(timecourse), label = "gNA")
plt.plot(timecourse, gK(timecourse), label = "gK")
plt.title('Ion currents over time')
plt.xlabel('Time in $ms$')
plt.ylabel('Ion current in $mA$')
plt.legend()
```

```python
def euler(f_func, x_0, t_max, dt):
    
    x = x_0
    nsteps = int(t_max / dt)
    xs = np.zeros(nsteps)
    xs[0] = x_0
    i = 1
    t = 0
    
    while i < nsteps:
        
        x = x + f_func(x,t)*dt
        xs[i] = x
        t += dt
        i += 1
    
    return xs

def myfunc(x):
    
    return -x * ( x + 1) * (x - 1) * (x - 1) * (x - 2)

np.vectorize(myfunc)
```
