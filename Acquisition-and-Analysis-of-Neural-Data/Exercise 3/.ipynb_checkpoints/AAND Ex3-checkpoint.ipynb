```python
#Part(b): PLot a historgram of the ISIs pooled across trials:

Pooling = NewType('Pooling', Union["Trials", "Windows"])

def ISI_pool (data, pooling: Optional[Pooling]):
    
    
    result = np.diff(data, axis=0).flatten() if data.shape[0] != 1 else np.diff(data).flatten()
    
    result2 = np.where(np.isnan(result), 0, result)
    
    #Compute CV and Fano Factor:
    
    cv, fano = np.round(CV(result2), decimals=4), np.round(Fano(result2),decimals=4)
    
    plt.figure(figsize = (10,5))
    plt.hist(result, color = 'tomato', linewidth = 15, alpha = .8, bins = 50, density = True)
    plt.title("Histogram of the interspike intervals, CV={}, Fano={}".format(cv,fano))
    plt.xlabel("ISIs in seconds")
    plt.ylabel("Value of Probability Density")
    
    return None


ISI_pool(spike_data, "Trials")
```

```python
windowSize = NewType('windowSize', float)

def windows(data, wS: windowSize):
    
    #Number of windows:
    
    num_windows = int(np.floor(data.shape[0]/wS))
    
    #Intialize return array:
    
    windows = np.zeros((num_windows, data.shape[1]))
    
    k = 0
    
    for col in range(windows.shape[1]):
        
        k = 0
        
        while k < num_windows:
            
            #Count the spikes within a window:
            
            windows[k, col] = len(list(filter(lambda x: k * wS <= x < (k+1) * wS, data[:,col])))
            
            k += 1
            
    if data.shape[1] != 1:

        fanoTime = [Fano(row) for row in windows]
        fanoTrials = [Fano(col) for col in windows.T]
        
    else:
        
        fanoTime = Fano(windows)
        fanoTrials = fanoTime
            
    return windows, fanoTime, fanoTrials

fig, axs = plt.subplots(1,2, figsize = (20,10))

colors = ['darkred', 'tomato', 'lightsalmon']

for wInd, wS in np.ndenumerate(np.array([.3,.5,1])):
    
    num_windows = int(np.floor(spike_data.shape[0]/wS))
    
    result = windows(spike_data,wS)
    
    #print("The Trial-resolved Fano factor for a windowsize {} is: {}".format(wS, result[2]))
    #print("The Time-resolved Fano factor for a windowsize {} is: {}".format(wS, result[1]))
    axs[0].plot(np.arange(100), result[2], c=colors[wInd[0]], label=wS)
    axs[0].set_xlabel('Trials')
    axs[0].set_ylabel('Fano factor')
    axs[0].set_title('Trial-resolution')
    axs[0].legend()
    axs[1].plot(np.arange(num_windows), result[1], c=colors[wInd[0]], label=wS)
    axs[1].set_xlabel("window number")
    axs[1].set_ylabel('Fano factor')
    axs[1].set_title('Time-resolution')
    axs[1].legend()
    
```

```python
#Get the spike trains:

def spike_train (spike_times, time_range):
    
    """
    Compute spike train:
    @params:
    spike_times: an array of spike events
    time_range: an array of times lasting covering the span 
    of the entire experiment at some temporal resolutio
    """
    
    result = np.zeros(time_range.shape[0])
    
    timeT = np.round(time_range, decimals=4)
    
    spikeT = np.where(np.isnan(spike_times), 0, spike_times)
    
    spikeT = np.round(spike_times, decimals=4)

    for tI, t in np.ndenumerate(timeT):
        
        dirac_t = np.where(np.array(t - spikeT) == 0, 1, 0)
        result[tI] = dirac_t.sum()
        
    return result

#Get the trial average:

def trial_average (spike_data, num_trials, time_range):
    
    """
    Compute the trial-averaged spike train:
    @params:
    spike_data: ----"----
    num_trials: The number of trials over which to take the average
    time_range: ----"----

    """
    
    result = np.zeros((num_trials, time_range.shape[0]))
    
    for trial in range(num_trials):
        
        result[trial,:] = spike_train(spike_data[:,trial], time_range)
    
    return result.mean(axis = 0)

#Compute the auto correlation:

def auto_cor (data, trial_times):
    
    tav = trial_average(data, data.shape[1], trial_times)
    
    result = np.correlate(tav, tav, "same")
    
    return result

spike_auto_correlation = auto_cor(spike_data, trial_times)
```

```python
#Use procedure from generatePoisson to generate homogenous Poisson spikeTrain:

def hom_poisson(n, rate):
    
    randNumbers = np.random.rand(n)
    ISIs = -np.log(randNumbers)/rate
    ISIs = ISIs*1000 #to ms
    
    return np.cumsum(ISIs)

#Parameters for the k-Gamma processes:

K = np.array([2,3,5,10])

'''
Generate matrix for the gamma processes: each row will contain
hundred spikes from one gamma process with k in K

'''

Gamma = np.zeros((len(K), 1000))

for kI, k in np.ndenumerate(K):
    
    
    #Generates a homogenous Poisson with k*n evenst keeping only each kth event:
    
    hom = hom_poisson(k*1000, 100)
    gamma = hom[0:k*1000:k]
    
    Gamma[kI,:] = gamma
    
    


#Part(b): Compute histograms and CVS:

plt.figure(figsize = (20,6))

colors = ['firebrick','darkred', 'tomato', 'lightsalmon', 'coral']

rowI = 0
for row in Gamma:
    
    color = colors[rowI]
    
    result = np.diff(row).flatten()
    
    cv, fano = np.round(CV(result), decimals=4), np.round(Fano(result),decimals=4)
    
    print('The CV for the {}th gamma process is: {}'.format(K[rowI], cv))
    print('The Fano factor for the {}th gamma process is: {}'.format(K[rowI], fano))
    
    plt.hist(result, color=color, linewidth = 15, alpha = 1 - 1 / K[rowI], bins = 50, density = True, label= K[rowI])
    plt.title("Histogram of the interspike intervals, CV={}, Fano={}".format(cv,fano))
    plt.xlabel("ISIs in ms")
    plt.ylabel("Value of Probability Density")
    plt.legend()
    
    rowI += 1
    
    
    


    
```

```python
from scipy.special import factorial

def pISI (t, k, rate=100):
    
    num = rate * (rate * t)**(k - 1) * np.exp(-rate * t)
    denom = factorial(k-1)
    
    return num / denom

```

```python
plt.figure(figsize = (20,6))

for kI, k in np.ndenumerate(K):
    
    time_range = np.arange(0,2,.001)
    
    psi_partial = lambda t: pISI(t,k)
    np.vectorize(psi_partial)
    #convert to ms and make into density:
    plt.plot(time_range*1000, psi_partial(time_range) / 1000, color=colors[kI[0]], label = k) 
    plt.xlim(0,250)
    plt.title(r'Analytical Solutions for $P_{ISI}$')
    plt.xlabel("ISIs in ms")
    plt.ylabel(r"Density of $P_{ISI}$")
    plt.legend()
```

Comparing the results we can see that they match fairly well. For higher $ks$, the distribution approaches a Gaussian curve, which is the thereotical resutl. The histogram should then look more and more like a binomial distribution, which it does.
