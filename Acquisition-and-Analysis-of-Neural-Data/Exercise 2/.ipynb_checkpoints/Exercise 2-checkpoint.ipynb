```python
import numpy as np
import matplotlib.pyplot as plt
import sys
from typing import Optional
```

```python
spike_data = np.loadtxt("SpikeTimes.dat.txt")
spike_data = spike_data / 1000 #Converting to seconds

m, n = spike_data.shape
TRIAL_LENGTH, dt = 5.5, .0001

trial_times = np.arange(0, TRIAL_LENGTH, dt)
```

```python
#Make an eventplot of the data:

def rasterplot (data, trial_length = TRIAL_LENGTH):
    
    plt.figure(figsize = (20,4))
    
    colors = ['C{}'.format(i) for i in range(10)]
    
    colors = np.repeat(colors, 10)
    
    plt.eventplot(data, color = 'black')
    plt.tick_params(labelsize=12)
    plt.xlim(0, trial_length)
    plt.ylim(0, 100)
    plt.xlabel("time in $s$")
    plt.ylabel("trial", fontsize=12);
    plt.title("Rasterplot of the spike data", fontsize = 20)
    
    
    plt.show()


rasterplot(spike_data.T)

#So what you do is: For every time in time_range take the sum of
#that time minus all the spike_times and write it into an array
#then average this across trials
```

So the spikes are organised in well-defined bands in which the spikes appear to be more or less uniformly distributed. When we calculate the firing rates we should thus see – given a large enough window size – a constant firing rate within those bands that quickly drops to zero when we go outside them.

```python
#Define the spike train:

def spike_train (spike_times, time_range):
    
    result = np.zeros(time_range.shape[0])
    
    timeT = np.round(time_range, decimals=4)
    
    spikeT = np.where(np.isnan(spike_times), 0, spike_times)
    
    spikeT = np.round(spike_times, decimals=4)

    for tI, t in np.ndenumerate(timeT):
        
        dirac_t = np.where(np.array(t - spikeT) == 0, 1, 0)
        result[tI] = dirac_t.sum()
        
    return result

def trial_average (spike_data, num_trials, time_range):
    
    
    result = np.zeros((num_trials, time_range.shape[0]))
    
    for trial in range(num_trials):
        
        result[trial,:] = spike_train(spike_data[:,trial], time_range)
    
    return result.mean(axis = 0)

#Rectangular window:

rec_window_func = lambda tau, dt: 1/dt if -dt/2 <= tau < dt/2 else 0

def convolution(window, spikes, wp: Optional):
    
    """
    Returns the convolution of window with spike times
    @params:
    window: the kind of window to be used
    wp: a window paramater that can be used to scale e.g.
    the width of the window 
    """
    
    w = np.array([window(time, wp) for time in trial_times])
    
    return np.convolve(spikes,w , 'full')
```

```python
#Plot the results:

def plot_tafr(data, \
              time_range = trial_times, \
              widths = np.array([.03, .09, .2]), \
              num_trials = np.array([10,50,100])):
    
    fig, axs = plt.subplots(3,1, figsize = (20,6))
    
    colors = ['darkred', 'tomato', 'lightsalmon']
    
    plt.tight_layout()
    
    for Itrial, trial in np.ndenumerate(num_trials):
        
        for Iwidth, width in np.ndenumerate(widths):
            
            tav = trial_average(data, trial, time_range)
            result = convolution(rec_window_func, tav, width)[:time_range.shape[0]]
        
            axs[Itrial].plot(time_range, result, \
                             linewidth = width * 100, c = colors[Iwidth[0]], \
                             alpha = 1 - (width * 3), \
                             label = "width = {}".format(width))
            axs[Itrial].set_xlabel("Time in $s$")
            axs[Itrial].set_ylabel("firing rate")
            axs[Itrial].set_title("Trial averaged firing rate for first {} trials".format(num_trials[Itrial[0]]))
            axs[Itrial].legend()
               
plot_tafr(spike_data)
        
```

The plots illustrate a smoothing effect that comes from two sources. Firstly, since the trials are independent, we can see that the variance decreases as we increase the number of trials we sample from (this is essentially the law of large numbers). Secondly, as we increase the window size the firing rates in an interval approach the temporal average of the firing rates in that interval and our firing rate estimates becomes smoother.
```python
from scipy.io import loadmat

matData = loadmat("STA_data.mat")

stimulus = matData["Stimulus"]

spike_data2 = matData["SpikeTimes"] / 1000
```

```python
TRIAL_LENGTH2, dt2 = 1, .0001
trial_times2 = np.arange(0, TRIAL_LENGTH2, dt2)
```

```python
def STA(data, trial_times, stimulus = stimulus):
    
    tav = trial_average(data, data.shape[1], trial_times)
    
    result = np.correlate(tav, stimulus, "same")
    
    return result

strigAv = STA(spike_data2, \
              trial_times2, \
              stimulus.reshape(10000))

def plot_STA ():

    plt.figure(figsize = (10,6))
    
    plt.xlim(50, 0) #time backwards
    plt.title("Spike triggered average")
    plt.plot(strigAv[5000:5500], c = 'tomato', label = "C(r$\tau$)") # 0 <= tau <= 50ms
    plt.xlabel("Time in ms")
    plt.ylabel("C(r$\tau$)")
    plt.legend()
    
    plt.show()
    
    plt.plot(strigAv)
```

```python
def plot_STA ():

    fig, axs = plt.subplots(1,2, figsize = (20,6))
    
    axs[0].plot(np.arange(0,1, dt2), strigAv, c = 'tomato')
    axs[0].set_xlim(0,1)
    axs[0].set_xlabel("Time in s")
    axs[0].set_ylabel("Correlation")
    axs[0].set_title("Correlation between stimulus and response")
    
    axs[1].set_xlim(.05, 0) #time backwards
    axs[1].set_title("Spike triggered average")
    axs[1].plot(np.arange(0,.05, dt2),strigAv[5000:5500] / 1000, c = 'tomato', label = "C(r$\tau$)") # 0 <= tau <= 50ms
    axs[1].set_xlabel("Time in s")
    axs[1].set_ylabel(r"C($\tau$)")
    axs[1].legend()
    
    plt.show()

plot_STA()
```
