## Exercise 1: Plotting ERPs  (2 points)

Plot the average ERPs corresponding to *target* and *nontarget* stimuli in the time interval -100 to 1000 ms for the channels Cz and PO7.
In order to make the curves look reasonable, a so called 'baseline correction' should be performed: for each channel and trial, calculate the average across the prestimulus time interval -100 to 0 ms. This results in one value per channel and trial. Subtract this 'baseline' value from each (single channel/single trial) time course. 
The function `baseline` is provided in the `bci_minitoolbox`. It is recommended to have a look at the code.

```python
def pca(data):
    
    #Solve the eigenvalue problem:
    
    evals, evecs = np.linalg.eig(np.cov(data.T))
    
    #Return data in new basis:
    #Since evecs is orthonormalbasis we have evecs.T = Inv(evecs):

    new_data = np.array([evecs.T.dot(point) for point in data])
    
    return evals, evecs
```

```python
chans = ['Cz', 'PO7']
ival = [-100, 1000]
ref_ival = [-100, 0]

chans = ['Cz', 'PO7']
time_points = [380, 320]

data= np.zeros((len(chans), cnt.shape[1]))


for i in range(data.shape[0]):
    data[i,:] = cnt[np.where(clab==chans[i])]

    
#For Target trials: 

def plot_erps(data=data, 
              fs=fs, 
              mrk_class=mrk_class, 
              mrk_pos=mrk_pos, 
              ival=ival,
              ref_ival=ref_ival,
              target=True):
    
    #Targets
    epoT, epo_tT= bci.makeepochs(data, fs, mrk_pos[mrk_class==0], ival)
    epoT = bci.baseline(epoT, epo_tT, ref_ival)
    Oz, PO7 = np.transpose(epoT, (1,0,2))
    Oz_temp = Oz.mean(axis=1)
    PO7_temp = PO7.mean(axis=1)
    lenTargets = len(Oz)
    
    #non-Targets
    epoNT, epo_tNT= bci.makeepochs(data, fs, mrk_pos[mrk_class==1], ival)
    epoNT = bci.baseline(epoNT, epo_tNT, ref_ival)
    Oz2, PO72 = np.transpose(epoNT, (1,0,2))
    Oz2_temp = Oz2.mean(axis=1)
    PO72_temp = PO72.mean(axis=1)
    lenNon = len(Oz2_temp)

    targets = np.linspace(ival[0], ival[1], lenTargets)
    nonTargets = np.linspace(ival[0], ival[1], lenNon)
    
    fig, axs = plt.subplots(2,1, figsize=(15,20), sharex=True)
    plt.tight_layout()
    
    axs[0].set_title('Channel Oz')
    axs[0].plot(targets, Oz_temp, label= 'Targets')
    axs[0].plot(targets, Oz2_temp, label='Non-targets')
    axs[0].set_ylabel('Potential in $\mu$V')
    axs[0].legend()
    axs[1].set_title('Channel PO7')
    axs[1].set_xlabel('Time')
    axs[1].set_ylabel('Potential in $\mu$V')
    axs[1].plot(targets, PO7_temp, label='Tagets')
    axs[1].plot(targets, PO72_temp, label='None-targets')
    axs[1].legend()
    
    targetData = np.zeros((Oz.shape[1], 2))
    nonTargetData = np.zeros((Oz2.shape[1], 2))
    targetData[:,0] = Oz[48]
    targetData[:,1] = PO7[42]
    targetMean = targetData.mean(axis=0)
    targetData -= targetMean #center
    nonTargetData[:,0] = Oz2[48]
    nonTargetData[:,1] = PO72[42]
    nTargetMean = nonTargetData.mean(axis=0)
    nonTargetData -= nTargetMean
    
    #Exercise 2:
    plt.figure(figsize=(10,10))
    
    plt.xlabel('Channel Cz at 380ms')
    plt.ylabel('Channel PO7 at 320ms')
    plt.scatter(targetData[:,0], targetData[:,1], alpha=.7, s=30, label='Targets')
    plt.scatter(nonTargetData[:,0],nonTargetData[:,1], alpha=.7,s=30, label='Non-targets')
    plt.legend()
    
    evals1, evecs1 = pca(targetData)
    evals2, evecs2 = pca(nonTargetData)
    
    #Draw PCAs of targets:
    
    x_pos, y_pos = targetMean
    x_direct, y_direct = evecs1.T[0,:]
    plt.quiver(x_pos, y_pos, x_direct, y_direct,\
               color= 'coral', label='Target PCAs', scale=np.sqrt(evals1[1]))
    x_direct, y_direct = evecs1.T[1,:]
    plt.quiver(x_pos, y_pos, x_direct, y_direct,\
               color='coral', scale=np.sqrt(evals1[0]))

    
    #Draw PCAs of non-target:
    
    x_pos, y_pos = nTargetMean 
    x_direct, y_direct = evecs2.T[0,:]
    plt.quiver(x_pos, y_pos, x_direct, y_direct,\
              color= 'silver', label='Non-target PCAs', scale=np.sqrt(evals2[1]))
    x_direct, y_direct = evecs1.T[1,:]
    plt.quiver(x_pos, y_pos, x_direct, y_direct,\
              color='silver', scale=np.sqrt(evals2[0]))
    
    
    #Calculate the ellipses for bonus part:
    
    unitCircle1 = np.zeros((100,2))
    unitCircle1[:,0] = np.cos(np.linspace(0,2*np.pi,100))
    unitCircle1[:,1] = np.sin(np.linspace(0,2*np.pi,100))
    unitCircle2 = unitCircle1.copy()
    
    
    D1, D2 = np.eye(2) * np.sqrt(evals1), np.eye(2) * np.sqrt(evals2)
    ellipse1 = evecs1 @ D1 @ evecs1.T @ unitCircle1.T
    ellipse1 = ellipse1.T
    ellipse1 += targetMean
    ellipse2 = evecs2 @ D2 @ evecs2.T @ unitCircle2.T
    ellipse2 = ellipse2.T
    ellipse2 += nTargetMean
    plt.plot(ellipse1[:,0], ellipse1[:,1], color='coral', linewidth=5)
    plt.plot(ellipse2[:,0], ellipse2[:,1], color='silver', linewidth=5)
    plt.legend()
    
    return None

plot_erps()
```

## Exercise 3: Scalp Topographies of ERPs  (2 points)

Calculate the average potential separately for the classes *target* and *non-target* in the following time intervals [ms]: [160 200], [230 260], [300 320], and [380 430]. (This involves averaging over all trials of the respective class and averaging over all sample points in the respective time interval.)
Visualize the result as scalp topographies using the function `bci.scalpmap`, i.e., 4 maps for each class.
**Note:** In order to make the maps look reasonable, do a *baseline correction* as in the previous task. To make the maps comparable, use a common scale for the colorbar (see help of `scalpmap`). For ERPs, a range should be chosen for the colormap, which is symmetric around 0.

```python
ivals = [[160, 200], [230, 260], [300, 320], [380, 430]]

def scalp_topo(data,
               ival=ival,
               fs=fs, 
               mrk_class=mrk_class, 
               mrk_pos=mrk_pos,
               ref_ival=ref_ival,
               target=True):
    
    classer = 0 if target==True else 1
    
    epochs, epochs_t = bci.makeepochs(data, fs, mrk_pos[mrk_class==classer], ival)
    epochs = bci.baseline(epochs, epochs_t, ref_ival)
    
    result = epochs.mean(axis=2)
    
    return result

def plot_scalpmaps(ivals=ivals):

    resultTarget = scalp_topo(cnt)
    resultNonTarget = scalp_topo(cnt, target=False)


    nrows = len(mrk_className)
    ncols = len(ivals)

    plt.figure(figsize=(10,5))
    plt.tight_layout(w_pad=5, h_pad=10)
    plt.suptitle('Scalpmap of Target ERPs')
    

    for ivalI, ival in enumerate(ivals):
        
        lower, upper = np.array([int(i*fs/1000) for i in ival]) + 10
        mapTarget = resultTarget.copy()
        mapNonTarget = resultNonTarget.copy()
        plt.subplot(nrows, ncols, ivalI+1)
        plt.title(ival)
        bci.scalpmap(
            mnt,
            mapTarget[lower:upper, :].mean(axis=0),
            clim = [-6,6],
            cb_label = 'Potential' if ivalI+1==4 else 0,
            contours = True
        )
    
    plt.figure(figsize=(10,5))
    plt.tight_layout(h_pad=5, w_pad=5)
    plt.suptitle('Scalpmap of Non-arget ERPs')
    for ivalI, ival in enumerate(ivals):
        lower, upper = np.array([int(i*fs/1000) for i in ival]) + 10
        mapNonTarget = resultNonTarget.copy()
        plt.subplot(nrows, ncols, ivalI+5)
        plt.title(ival)
        bci.scalpmap(
            mnt,
            mapNonTarget[lower:upper, :].mean(axis=0),
            clim = [-6,6],
            cb_label = 'Potential' if ivalI+1==4 else 0,
            contours=True
        )
    
    return None
    
plot_scalpmaps()


```

We can clearly see that in the target cases the event related potential are much higher. But we can also see that there appears to be some oscillations in the ERPs for the target stimuli.

## Exercise 4: Visualization with the Biserial Correlation Coefficient  (3 points)

Implement a function for the calculation of the signed `r^2`-value (see point-biserial correlation coefficient in BCI lecture \#02). From the given data set, extract epochs for the time interval [-100 600] ms relativ to each stimulus presentation.  Calculate for each channel and each point in time the signed `r^2`-value wrt. classes
*target* vs. *non-target* and visualize this (channel `x` time points) matrix (`pl.imshow`). Again, use 100 ms prestimulus for baseline correction.

```python
def signed_r_square(epo, y):
    '''
    Synopsis:
        epo_r = signed_r_square(epo, y)
    Arguments:
        epo:    3D array of segmented signals (time x channels x epochs), 
                see makeepochs
        y:      labels with values 0 and 1 (1 x epochs)
    Output:
        epo_r:  2D array of signed r^2 values (time x channels)
    '''
    
    targetM = epo[:,:,y==0].mean(axis=2)
    nonTargetM = epo[:,:,y==1].mean(axis=2)
    
    factor = len(targetM)*len(nonTargetM) / ((len(targetM)+len(nonTargetM)**2))
    mean_dist = targetM - nonTargetM
                                             
    r2 = factor * mean_dist / epo.var(axis=2)
                                             
    return r2 * np.sign(mean_dist)                                         
```

```python
ival = [-100, 600]
ref_ival = [-100, 0]

def plot_biserial (ival=ival, ref_ival=ref_ival):
    
    epo, epo_t = bci.makeepochs(cnt, fs, mrk_pos, ival)
    epo = bci.baseline(epo, epo_t, ref_ival)
    epo_r = signed_r_square(epo, mrk_class)
    
    fig, ax0 = plt.subplots(figsize=(10,10))
    plt.title('Point biserial correlation coefficient')
    im1 = ax0.imshow(epo_r.T, cmap='rainbow')
    divider = make_axes_locatable(ax0)
    cax = divider.append_axes("right", size="5%", pad=0.2)
    plt.colorbar(im1, cax=cax)
    ax0.set_ylabel('Channels')
    ax0.set_xlabel('time in $ms$')
    ax0.set_xticks(np.arange(0, 70+1, 10))
    ax0.set_xticklabels(np.arange(-100, 601, 100))
    ax0.set_yticks(np.arange(0, 55, 10),
    ax0.set_yticklabels(np.array([lab for lab in clab])[::9])
)
    
    return None

plot_biserial()
    
    
```
