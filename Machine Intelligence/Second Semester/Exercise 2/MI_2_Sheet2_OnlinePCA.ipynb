```python
def pca(data):
    
    #Solve the eigenvalue problem:
    
    evals, evecs = np.linalg.eig(np.cov(data.T))
    
    sort = np.zeros((evecs.shape[0]+1,evecs.shape[1]))
    
    sort[0,:] = evals
    sort[1:,:] = evecs
    
    sort = sort[:,sort[0].argsort()]
    
    evals = sort[0,:]
    evecs = sort[1:,:] 
    
    #Return data in new basis:
    #Since evecs is orthonormalbasis we have evecs.T = Inv(evecs):

    new_data = np.array([evecs.T.dot(point) for point in data])
    
    return evals, evecs, new_data
```

```python
# PCA for each Block 

x = 0

pc_matrix = np.zeros((10,2))

for counter, value in enumerate(np.arange(200, 2200, 200)):
    
    eivals, eivecs, conv_data = pca(npdata[x:value,1:])
    pc_matrix[counter,:] = eivecs[0,:]

```

```python
# plot of the first PC of each block together with original data (version 1 of presentation)
origin = [0], [0]
data.plot.scatter( x="V1",y="V2", c = "Block",colormap='viridis')
plt.quiver(*origin, pc_matrix[:,0],pc_matrix[:,1],  color=['r','b','g','c','m','y','k','r','b','g'], scale = 10)
plt.suptitle('Datapoints by blocks')
plt.xlabel('V1')

plt.show()
```

```python
def OjasRule(initialweight,learningstep, data):
    y = initialweight.T @ data 
    
    w_new = initialweight + learningstep*y*(data-y*initialweight)
    
    return w_new
```

```python
weights = np.zeros((2000,2))
weight_0 = np.random.normal(1, 0.01, (2))
epsilon = 0.002
print(weight_0)

for value in np.arange(0,2000,1):
    
    weights[value,:] = OjasRule(weight_0,epsilon,npdata[value,1:3])
    
  
```

```python
weight_plot = np.zeros((2000,3))
weight_plot[:,0] = npdata[:,0]
weight_plot[:,1:] =weights
```

```python
weightdf = pd.DataFrame(weight_plot, columns = ['Block', 'W1', 'W2'])
```

```python
data.plot.scatter( x="V1",y="V2", c = "Block",colormap='viridis')
weightdf.plot.scatter( x="W1",y="W2", c = "Block",colormap='viridis')
plt.suptitle('Original Datapoints by blocks and weights with epsilon = 0.002')
plt.xlabel('V1')

plt.show()
```
