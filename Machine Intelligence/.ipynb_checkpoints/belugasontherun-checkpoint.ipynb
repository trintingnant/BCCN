```python
import numpy as np
import os
import pandas as pd
import functools
import math




# Exercise 1a:

os.getcwd()

data = np.loadtxt(fname= 'applesOranges.csv', delimiter = ',', skiprows = 1)                  

import matplotlib.pyplot as plt 

X, Y, labels = data[:,0], data[:,1], data[:,2]

oranges = labels > 0.5
apples = labels < 0.5

plt.title('Scatter Plot of x1 vs x2')
scatter_1 = plt.scatter(data[oranges][:,0],data[oranges][:,1], c = 'orange', label = 'oranges')
scatter_2 = plt.scatter(data[apples][:,0],data[apples][:,1], c = 'green', label = 'apples')
legend1 = plt.legend((scatter_1,scatter_2), ('oranges', 'apples'), loc="lower left", title="Classes")
plt.xlabel('x1')
plt.ylabel('x2')


#Exercise 1bi:

theta = 0

gamma = np.radians(np.arange(0,190,10))

x_coords = np.cos(gamma)
y_coords = np.sin(gamma)



#Compute the performance for one vector:

def performance(w1, w2, theta):
    
    percentage = []
    
    res = (w1 * X + w2 * Y - theta > 0) 
    
    percentage = np.sum(np.equal(res, labels)) / (len(labels))
        
    return percentage



#Get the performances of all the vectors:

def performance_all (theta):

    perf = []
    
    for x_coord, y_coord in zip(x_coords,y_coords):
    
        perf = perf + [performance(x_coord,y_coord, theta)]
        
    return perf


best_vector = np.argmax(performance_all(0))

#Exercise 1bii:

gamma_2 = np.arange(0,190,10)

plt.figure()
plt.title('Plot a curve showing  as a function of ')
plt.plot(gamma_2, performance_all(0))
plt.xlabel('gamma = angle between weight vector and the x-axis')
plt.ylabel('psi = % correct classifications')


#Exercise 1c:

vary_bias = np.arange(-3,3,0.1)

#best weights

w_x = x_coords[best_vector]
w_y = y_coords[best_vector]

w_performance_partial = lambda x: performance(w1 = w_x, w2 = w_y, theta = x)


best_theta = vary_bias[np.argmax(list(map(w_performance_partial, vary_bias)))]


#Exercise 1d:

def classify (w_1, w_2, theta):
    
    temp_labels = np.zeros(len(labels))
    
    for i in range (len(X)):
        
        if ((w_1 * X[i] + w_2 * Y[i] - theta) > 0):
        
            temp_labels[i] = 1 
        
        else:
            
            temp_labels[i] = 0 
        
    return temp_labels

classes = (classify(w_x, w_y, best_theta))

yes = classes < 0.5
no = classes > 0.5


plt.figure()

plt.title('Scatter Plot of x1 vs x2 and generated data points')
scatter_3= plt.scatter(data[yes][:,0],data[yes][:,1], c = 'green')
scatter_4= plt.scatter(data[no][:,0],data[no][:,1], c = 'orange')
legend1 = plt.legend((scatter_3,scatter_4), ('apples', 'oranges'), loc="lower left", title="Classes")
plt.xlabel('x1')
plt.ylabel('x2')

#Plot the vector:

xpts = np.linspace(-2,2,100)
ypts = np.linspace(-2,2,100)

func = lambda x,y: x * X[best_vector] + y * Y[best_vector]

funct = np.vectorize(func) 

plt.plot(xpts, funct(xpts,ypts))

#Exercise 1e:



#Create matrix for plotting:

matr = np.zeros((len(x_coords), len(vary_bias)))

for i in range (len(x_coords)):
    for j in range (len(vary_bias)):
    
        matr[i,j] = performance(x_coords[i], y_coords[i], vary_bias[j])
    

plt.figure()

plt.title("Heatmap of best results")
plt.imshow(matr, cmap = 'hot', interpolation = 'nearest')


```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import os 
import random 

#H2

#Exercise 2a:

#Compute the output:

def mlp_out(w21, w10, b, x):
    
    first_layer = np.tanh(w10 * x - b)
    second_layer = np.sum(w21 * first_layer)
    
    return second_layer

#Create $num mlps:

def make_mlps (num):
    
    MLPS = []
    
    for i in range (num):
        
        MLPS = MLPS + [initialize_mlp()]
        
    return MLPS

#Initializes an mlp with the respective weights and biases:


def initialize_mlp ():
    
    w21 = np.array([np.random.normal(0,1) for i in range(10)])
    w10 = np.array([np.random.normal(0,2) for i in range (10)])
    b = np.array([np.random.uniform(-2,2) for i in range (10)])
    

    return {'w21': w21, 'w10': w10, 'b': b}
 
# Exercise 2b: Plot the input-ouput function for x \in [-2,2]



xpts = np.linspace(-2,2,100)


#Run an mlp on all possible input values:

def run_mlp (mlp):
    
    ypts = []
    
    for x in xpts:
        
        ypts = ypts + [mlp_out(mlp['w21'], mlp['w10'], mlp['b'], x)]
        
    
    return ypts 

# Plotting:


for i in range(50):
        
    plt.plot(xpts, run_mlp(initialize_mlp()))
    plt.title('Input Output Functions')
    plt.xlabel('x')
    plt.ylabel('y(x)')


    
#Same as before:


def initialize_mlp_new ():
    
    w21 = np.array([np.random.normal(0,1) for i in range(10)])
    w10 = np.array([np.random.normal(0,0.5) for i in range (10)])
    b = np.array([np.random.uniform(-2,2) for i in range (10)])
    

    return {'w21': w21, 'w10': w10, 'b': b}


def make_mlps_new (num):
    
    MLPS = []
    
    for i in range (num):
        
        MLPS = MLPS + [initialize_mlp()]
        
    return MLPS

plt.figure()

for i in range(50):
        
    plt.plot(xpts, run_mlp(initialize_mlp_new()))
    plt.title('Input Output Functions')
    plt.xlabel('x')
    plt.ylabel('y(x)')
    


#Exercise 2d:


MLP_1 = make_mlps(50)

MLP_2 = make_mlps(50)
                  
                  
                  
    
```

```python
from functools import partial


#Compute the Euclidean distance


def MSE (a,b):

    dist = sum([(x - y)**2 for x,y in zip(a,b)])
    
    return dist

g = lambda x: -x

#Get all mlp outputs over the whole range of inputs:


def get_ycoords (mlps):
    
    mlp_ypts = []
            
    for mlp in mlps:
        
        partial_mlp = lambda x: mlp_out(w21 = mlp['w21'], w10 = mlp['w10'], b = mlp['b'], x = x)
            
        mlp_ypts = mlp_ypts + [list(map(partial_mlp, xpts))]
        
    return mlp_ypts

#Compute the distances (g, mlp) for all the mlps 
    
    
def distance (mlps):
    
    f_ypts = np.linspace(2,-2,100)
    
    dist = []
    
    for y_coord in get_ycoords(mlps):
        
        dist = dist + [MSE(y_coord, f_ypts)]
        
    return dist

#Testing functionality; plotting two best functions

print(np.argmin(distance(MLP_1)))
print(np.argmin(distance(MLP_2)))

best_1 = MLP_1[np.argmin(distance(MLP_1))]
best_2 = MLP_2[np.argmin(distance(MLP_2))]

#print(best_1)
#print(best_2)

plt.plot(xpts,run_mlp(best_1))
plt.plot(xpts, run_mlp(best_2))
plt.title('Input Output Functions')
plt.xlabel('x')
plt.ylabel('y(x)')


```
